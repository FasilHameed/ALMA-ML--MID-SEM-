**Problem Statement:**

- The dataset provided comprises various medical predictor (independent) variables and one target (dependent) variable, known as "Outcome." 

- These independent variables include the number of pregnancies, glucose levels, blood pressure, skin thickness, insulin levels, BMI, diabetes pedigree function, and age.

- The objective is to develop a predictive model that can accurately predict the likelihood of a patient having diabetes based on these medical attributes. 

- This predictive model can aid healthcare professionals in early diagnosis and intervention, thereby improving patient outcomes and quality of care.

- The dataset contains information on 768 patients, with the "Outcome" variable indicating whether a patient has diabetes (1) or not (0). 

- With this information, the goal is to build a machine learning model capable of accurately classifying patients into these two categories based on their medical attributes.

- By leveraging machine learning algorithms, we aim to create a reliable and efficient tool for diabetes prediction, facilitating proactive healthcare management and personalized treatment strategies for individuals at risk of or already affected by diabetes.


# <div style="padding:18px;color:white;margin:0;font-size:24px;text-align:left;display:fill;border-radius:20px;background-color:#74acbc;overflow:hidden">5 | Handling Outliers (Z Score)</div>

```python
from scipy.stats import zscore

def handle_outliers_zscore(data, column, threshold=3):
    """
    Function to handle outliers in a numerical column using z-score method.

    Parameters:
    - data: DataFrame containing the data
    - column: Name of the numerical column to handle outliers
    - threshold: Z-score threshold for identifying outliers (default is 3)

    Returns:
    - DataFrame with outliers replaced by the median of the column
    """
    # Calculate z-scores for the column
    z_scores = zscore(data[column])

    # Identify outliers based on the threshold
    outliers = (z_scores > threshold) | (z_scores < -threshold)

    # Replace outliers with the median of the column
    data.loc[outliers, column] = data[column].median()

    return data

# Get a list of numerical columns
numerical_columns = df.select_dtypes(include=[np.number]).columns.tolist()

# Iterate over each numerical column and handle outliers using z-score method
for column in numerical_columns:
    df = handle_outliers_zscore(df, column)
```


# <div style="padding:18px;color:white;margin:0;font-size:24px;text-align:left;display:fill;border-radius:20px;background-color:#74acbc;overflow:hidden">6 | Smote resampling</div>

```python
# Splitting the dataframe into features (X) and target variable (y)
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# Applying SMOTE (Synthetic Minority Over-sampling Technique) to handle class imbalance
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Creating a new dataframe with resampled data
df_resampled = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='Outcome')], axis=1)

# Plotting the distribution of the target variable after SMOTE
plt.figure(figsize=(6, 4))
sns.countplot(x='Outcome', data=df_resampled)
plt.title('Distribution of Outcome (After SMOTE)')
plt.show()
```


# <div style="padding:18px;color:white;margin:0;font-size:24px;text-align:left;display:fill;border-radius:20px;background-color:#74acbc;overflow:hidden">7 | Random Forest</div>

```python

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)


rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)

rf_accuracy = accuracy_score(y_test, rf_pred)

rf_report = classification_report(y_test, rf_pred, target_names=['Class 0', 'Class 1'], output_dict=True)

plt.figure(figsize=(2, 1))
sns.heatmap([[rf_accuracy]], annot=True, fmt=".2%", cmap='Blues', cbar=False, linewidths=.5, xticklabels=False, yticklabels=False)
plt.title('Accuracy for Random Forest')
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(pd.DataFrame(rf_report).iloc[:-1, :].T, annot=True, cmap='Blues', linewidths=.5)
plt.title('Classification Report for Random Forest')
plt.show()
```

# Ploting metrics For Comparison

```python
import matplotlib.pyplot as plt

def plot_comparison(train_df, test_df, metric='accuracy'):
    models = train_df.index.tolist()  # Assuming the models are in the index
    train_scores = train_df[metric].values
    test_scores = test_df[metric].values
    
    plt.figure(figsize=(10, 6))
    plt.bar(models, train_scores, color='blue', alpha=0.5, label='Train')
    plt.bar(models, test_scores, color='orange', alpha=0.5, label='Test')
    plt.title(f'Comparison of {metric} for Different Models')
    plt.xlabel('Models')
    plt.ylabel(metric.capitalize())  # Capitalize the metric for better display
    plt.xticks(rotation=45)
    plt.legend()
    plt.show()

# Plot comparison of accuracy
plot_comparison(train_classification_df, test_classification_df, metric='accuracy')

# Plot comparison of other metrics (e.g., precision, recall, f1-score)
plot_comparison(train_classification_df, test_classification_df, metric='precision')
plot_comparison(train_classification_df, test_classification_df, metric='recall')
plot_comparison(train_classification_df, test_classification_df, metric='f1-score')
```
